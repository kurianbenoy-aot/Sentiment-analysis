```python
new_df = pd.DataFrame({'ID': df['ID'],
                   'STAFF': staffs,
                   'LOCATIONS': locations,
                   'SERVICES': services,
                    'OVERALL SENTIMENT': overall,
                    'LocationId': df['locationID'],
                    'FeedbackID': df['ID'],
                    'ServiceID': df['ServiceID']})
```
